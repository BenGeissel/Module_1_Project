```python
tmdb_budgets_df = pd.read_csv('tn.movie_budgets.csv')
tmdb_budgets_df.head()
```

```python
# tmdb_budgets_df.info()
# tmdb_budgets_df.describe()
tmdb_budgets_df.isnull().sum()
```

```python
tmdb_budgets_df['movie_date'] = tmdb_budgets_df.movie + ' ' + tmdb_budgets_df.release_date
```

```python
tmdb_budgets_df.production_budget = tmdb_budgets_df.production_budget.apply(lambda x: x.replace('$', ''))
tmdb_budgets_df.production_budget = tmdb_budgets_df.production_budget.apply(lambda x: x.replace(',', ''))
tmdb_budgets_df.production_budget = tmdb_budgets_df.production_budget.astype('int')

tmdb_budgets_df.domestic_gross = tmdb_budgets_df.domestic_gross.apply(lambda x: x.replace('$', ''))
tmdb_budgets_df.domestic_gross = tmdb_budgets_df.domestic_gross.apply(lambda x: x.replace(',', ''))
tmdb_budgets_df.domestic_gross = tmdb_budgets_df.domestic_gross.astype('int')

tmdb_budgets_df.worldwide_gross = tmdb_budgets_df.worldwide_gross.apply(lambda x: x.replace('$', ''))
tmdb_budgets_df.worldwide_gross = tmdb_budgets_df.worldwide_gross.apply(lambda x: x.replace(',', ''))
tmdb_budgets_df.worldwide_gross = tmdb_budgets_df.worldwide_gross.astype('int')
```

```python
fig = plt.figure(figsize = (8,5))
sns.distplot(tmdb_budgets_df.production_budget, bins = 50, color = 'darkred', kde = True);
plt.xlabel("Budget in 100,000,000's")
plt.ylabel('Distribution Amount')
plt.title('Budget Spread', fontsize = 20)
plt.show()
fig.savefig('budgetspread_tmdb.png')
#Many smaller budgets than large budgets
```

```python
tmdb_budgets_df['binned_budget'] = pd.cut(x=tmdb_budgets_df['production_budget'],
                                          bins= [0, 500000, 1000000, 5000000, 10000000, 25000000, 50000000, 100000000, 500000000])
```

```python
tmdb_budgets_df['Gross_ROI'] = ((tmdb_budgets_df.domestic_gross - tmdb_budgets_df.production_budget) / tmdb_budgets_df.production_budget)
tmdb_budgets_df['Worldwide_ROI'] = ((tmdb_budgets_df.worldwide_gross - tmdb_budgets_df.production_budget) / tmdb_budgets_df.production_budget)
```

```python
tmdb_budgets_df.describe()
```

```python
tmdb_budgets_df = tmdb_budgets_df[tmdb_budgets_df.domestic_gross != 0]
tmdb_budgets_df = tmdb_budgets_df[tmdb_budgets_df.worldwide_gross != 0]
# Data Assumption: Only movies that actually go to the box office. Not straight to Netflix or DVD because they would have 0 gross
```

```python
tmdb_budgets_domestic_pivot_df = pd.pivot_table(tmdb_budgets_df, index = 'movie_date',
                                              columns = 'binned_budget',
                                              values = 'Gross_ROI',
                                               aggfunc = 'sum')
```

```python
tmdb_budgets_domestic_pivot_df.head(15)
```

```python
tmdb_budgets_worldwide_pivot_df = pd.pivot_table(tmdb_budgets_df, index = 'movie_date',
                                              columns = 'binned_budget',
                                              values = 'Worldwide_ROI',
                                               aggfunc = 'sum')
```

```python
tmdb_budgets_worldwide_pivot_df.head(15)
```

```python
tmdb_budgets_domestic_pivot_df.count(axis = 0)
```

```python
domestic_roi_by_budget = tmdb_budgets_domestic_pivot_df.mean()
domestic_roi_by_budget
```

```python
tmdb_budgets_worldwide_pivot_df.count(axis = 0)
```

```python
worldwide_roi_by_budget = tmdb_budgets_worldwide_pivot_df.mean()
worldwide_roi_by_budget
```

```python
worldwide_roi_dict = {}
for label in tmdb_budgets_worldwide_pivot_df.columns:
    worldwide_roi_dict[str(label)] = worldwide_roi_by_budget[label]
worldwide_roi_dict
```

```python
domestic_roi_dict = {}
for label in tmdb_budgets_domestic_pivot_df.columns:
    domestic_roi_dict[str(label)] = domestic_roi_by_budget[label]
domestic_roi_dict
```

```python
fig = plt.figure(figsize=(20, 10))
X = np.arange(len(domestic_roi_dict))
ax = plt.subplot(111)
ax.set_facecolor('lightgray')
ax.bar(X, domestic_roi_dict.values(), width=0.3, color='b')
ax.bar(X+0.3, worldwide_roi_dict.values(), width=0.3, color='g', alpha = .7)
ax.legend(('Domestic','Worldwide'), fontsize = 20)
plt.xticks(X, domestic_roi_dict.keys(), fontsize = 12)
plt.title("ROI by Budget", fontsize=40)
plt.ylabel('ROI', fontsize = 25)
plt.xlabel('Budget', fontsize = 25)
plt.tight_layout()
plt.show()
fig.savefig('tmdb_roi_budget.png')
```

```python
bom_df = bom_df[bom_df.domestic_gross != 0]
bom_df.foreign_gross.fillna(value = 0, inplace = True)
bom_df.foreign_gross = bom_df.foreign_gross.astype('str')
bom_df.foreign_gross = bom_df.foreign_gross.apply(lambda x: x.replace(',', ''))
bom_df.foreign_gross = bom_df.foreign_gross.astype('float64')
```

```python
bom_df.dropna(inplace = True)
```

```python
bom_studio_grouped_df = bom_df.groupby(by = 'studio', axis = 0).mean()
```

```python
bom_studio_grouped_df
```

```python
bom_studio_grouped_topten_df = bom_studio_grouped_df.sort_values('domestic_gross', ascending = False)[:10]['domestic_gross']
```

```python
fig = plt.figure(figsize = (10,10), facecolor = 'w')
ax = fig.add_subplot(1,1,1)
ax.set_facecolor('lightblue')
bom_studio_grouped_topten_df.plot(kind = 'bar', color = 'black');
plt.xlabel('Studio', fontsize = 15)
plt.ylabel("Mean Domestic Gross in 100,000,000's", fontsize = 15)
plt.title('Top 10 Mean Domestic Gross by Studio', fontsize = 20)
fig.savefig('bom_domestic_gross_studio.png')
```
